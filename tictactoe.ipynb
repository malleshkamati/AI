```python
import random

winning = [[0, 1, 2], [3, 4, 5], [6, 7, 8], [0, 4, 8], [2, 4, 6], [0, 3, 6], [1, 4, 7], [2, 5, 8]]
learning_rate = 0.2
discount_factor = 0.9

def check_win(ele):
    if len(ele) < 3:
        return False
    for state in winning:
        for i in range(len(state) - 2):
            if ele[i] in state:
                if ele[i + 1] in state:
                    if ele[i + 2] in state:
                        return True
    return False

# Update Q-values for states after the game ends
def update_q_values(states, result):
    # Start from the last state and propagate the result back
    for i in reversed(range(len(states))):
        state = states[i]
        # The reward for the final state is the game result (1 for win, -1 for loss, 0 for draw)
        if i == len(states) - 1:
            states[state] = states.get(state, 0) + learning_rate * (result - states.get(state, 0))
        else:
            # Update based on the next state's value and propagate backward
            next_state = states[i + 1]
            future_value = states.get(next_state, 0)
            states[state] = states.get(state, 0) + learning_rate * (discount_factor * future_value - states.get(state, 0))

# Main play function
def play():
    board = [' '] * 9
    p1_moves = []  # Player 1 moves
    p2_moves = []  # Player 2 moves
    available = list(range(9))  # Available positions
    count = 0
    p1state = {}
    p2state = {}

    p1_q_states = []
    p2_q_states = []

    while True:
        print(board)
        if count % 2 == 0:
            # Player 1's move
            a = random.choice(available)
            board[a] = 'X'
            p1_moves.append(a)
            available.remove(a)

            # Track Q-states
            p1_q_states.append(tuple(p1_moves))

            # Initialize state if not yet present
            p1state.setdefault(tuple(p1_moves), 0)

            if check_win(p1_moves):
                print("Player 1 wins!")
                update_q_values(p1state, 1)  # Update for win
                update_q_values(p2state, -1)  # Update for loss
                print(f"p1state: {p1state}")
                print(board)
                return 1
        else:
            # Player 2's move
            a = random.choice(available)
            board[a] = 'O'
            p2_moves.append(a)
            available.remove(a)

            # Track Q-states
            p2_q_states.append(tuple(p2_moves))

            # Initialize state if not yet present
            p2state.setdefault(tuple(p2_moves), 0)

            if check_win(p2_moves):
                print("Player 2 wins!")
                update_q_values(p1state, -1)  # Update for loss
                update_q_values(p2state, 1)   # Update for win
                print(f"p2state: {p2state}")
                print(board)
                return -1

        count += 1

        # Check for a draw
        if count == 8:
            print("It's a draw!")
            update_q_values(p1state, 0)  # Update for draw
            update_q_values(p2state, 0)  # Update for draw
            print(f"p1state: {p1state}")
            print(f"p2state: {p2state}")
            return 0

        print(board)
```

```python
import math

class TicTacToe:
    def __init__(self):
        self.board = [' ' for _ in range(9)]
        self.current_winner = None

    def print_board(self):
        for row in [self.board[i*3:(i+1)*3] for i in range(3)]:
            print('| ' + ' | '.join(row) + ' |')

    def available_moves(self):
        return [i for i, spot in enumerate(self.board) if spot == ' ']

    def empty_squares(self):
        return ' ' in self.board

    def num_empty_squares(self):
        return self.board.count(' ')

    def make_move(self, square, letter):
        if self.board[square] == ' ':
            self.board[square] = letter
            if self.winner(square, letter):
                self.current_winner = letter
            return True
        return False

    def winner(self, square, letter):
        row_ind = square // 3
        row = self.board[row_ind*3 : (row_ind + 1) * 3]
        if all([spot == letter for spot in row]):
            return True

        col_ind = square % 3
        column = [self.board[col_ind+i*3] for i in range(3)]
        if all([spot == letter for spot in column]):
            return True

        if square % 2 == 0:
            diagonal1 = [self.board[i] for i in [0, 4, 8]]
            if all([spot == letter for spot in diagonal1]):
                return True
            diagonal2 = [self.board[i] for i in [2, 4, 6]]
            if all([spot == letter for spot in diagonal2]):
                return True

        return False

def alpha_beta_pruning(game, depth, alpha, beta, maximizing_player):
    if depth == 0 or game.current_winner:
        if game.current_winner == 'X':
            return 1
        elif game.current_winner == 'O':
            return -1
        else:
            return 0

    if maximizing_player:
        max_eval = -math.inf
        for move in game.available_moves():
            game.make_move(move, 'X')
            eval = alpha_beta_pruning(game, depth - 1, alpha, beta, False)
            game.board[move] = ' '
            game.current_winner = None
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha:
                break
        return max_eval
    else:
        min_eval = math.inf
        for move in game.available_moves():
            game.make_move(move, 'O')
            eval = alpha_beta_pruning(game, depth - 1, alpha, beta, True)
            game.board[move] = ' '
            game.current_winner = None
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)
            if beta <= alpha:
                break
        return min_eval

def get_best_move(game):
    best_score = -math.inf
    best_move = None
    for move in game.available_moves():
        game.make_move(move, 'X')
        score = alpha_beta_pruning(game, 5, -math.inf, math.inf, False)
        game.board[move] = ' '
        game.current_winner = None
        if score > best_score:
            best_score = score
            best_move = move
    return best_move

def play():
    game = TicTacToe()
    while game.empty_squares():
        game.print_board()
        if game.num_empty_squares() % 2 == 1:
            move = get_best_move(game)
            game.make_move(move, 'X')
            print(f"AI moved to square {move}")
        else:
            valid_square = False
            while not valid_square:
                square = input("Enter your move (0-8): ")
                try:
                    square = int(square)
                    if square not in game.available_moves():
                        raise ValueError
                    valid_square = True
                except ValueError:
                    print("Invalid square. Try again.")
            game.make_move(square, 'O')

        if game.current_winner:
            game.print_board()
            print(f"{game.current_winner} wins!")
            return

    game.print_board()
    print("It's a tie!")

if __name__ == "__main__":
    play()
```

```python
import math

# Initialize the board
board = [' ' for _ in range(9)]

# printing the board
def print_board(board):
    for i in range(0, 9, 3):
        print('|'.join(board[i:i+3]))

# Function to check if a player has won the game.
def check_win(board, player):
    win_conditions = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # Rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # Columns
        [0, 4, 8], [2, 4, 6]  # Diagonals
    ]
    # Check if the player occupies all positions in any of the win conditions.
    return any(all(board[i] == player for i in condition) for condition in win_conditions)

# Function to check if the board is full (no available moves left).
def is_full(board):
    return ' ' not in board  # If there are no empty spaces it will return True.

# Function to get a list of available moves (indices of empty spaces).
def get_available_moves(board):
    # Return a list of indices where the board is still empty (' ').
    return [i for i, spot in enumerate(board) if spot == ' ']

# Alpha-beta pruning algorithm for AI to choose the best move.
# It alternates between maximizing (AI's move) and minimizing (human's move).
def alpha_beta_pruning(board, depth, is_maximizing, alpha, beta):
    # Base cases: if 'O' wins, return +1, if 'X' wins, return -1, if tie, return 0.
    if check_win(board, 'O'):
        return 1
    if check_win(board, 'X'):
        return -1
    if is_full(board):
        return 0

    if is_maximizing:
        # Maximize the score for AI's move ('O').
        best_score = -math.inf
        for move in get_available_moves(board):
            # Try a move and call alpha-beta recursively.
            board[move] = 'O'
            score = alpha_beta_pruning(board, depth + 1, False, alpha, beta)
            # Undo the move (backtrack).
            board[move] = ' '
            best_score = max(score, best_score)  # Keep track of the best score.
            alpha = max(alpha, best_score)  # Update alpha (maximum lower bound).
            if beta <= alpha:  # Prune the branch if beta is less than or equal to alpha.
                break
        return best_score
    else:
        # Minimize the score for human's move ('X').
        best_score = math.inf
        for move in get_available_moves(board):
            board[move] = 'X'
            score = alpha_beta_pruning(board, depth + 1, True, alpha, beta)
            board[move] = ' '  # Undo the move (backtrack).
            best_score = min(score, best_score)  # Keep track of the best score.
            beta = min(beta, best_score)  # Update beta (minimum upper bound).
            if beta <= alpha:  # Prune the branch if beta is less than or equal to alpha.
                break
        return best_score

# Function to determine the best move for AI using alpha-beta pruning.
def get_best_move(board):
    best_score = -math.inf  # Start with the worst possible score.
    best_move = None  # Store the best move's index.
    alpha = -math.inf  # Initialize alpha (maximum lower bound).
    beta = math.inf  # Initialize beta (minimum upper bound).
    
    # Iterate over all available moves.
    for move in get_available_moves(board):
        board[move] = 'O'  # Try AI's move.
        score = alpha_beta_pruning(board, 0, False, alpha, beta)  # Call the pruning function.
        board[move] = ' '  # Undo the move (backtrack).
        if score > best_score:
            best_score = score  # Update the best score.
            best_move = move  # Update the best move.
    return best_move  # Return the best move index.

# Main function to play the Tic-Tac-Toe game.
def play_game():
    print("Welcome to Tic Tac Toe!")
    print_board(board)  # Display the initial empty board.

    # Game loop continues until there is a winner or a tie.
    while True:
        # Human's turn.
        print("Your turn")    
        human_move = int(input("Enter your move (0-8): "))  # Ask for the player's move.
        # Ensure the move is valid (within bounds).
        while human_move >= 9 or board[human_move] != ' ':
            print("Invalid move. Try again.")
            human_move = int(input("Enter your move (0-8): "))
        
        board[human_move] = 'X'  # Place 'X' on the board for the human's move.
        print_board(board)  # Print the updated board.

        # Check if the human has won.
        if check_win(board, 'X'):
            print("You win!")
            break
        # Check if the board is full, indicating a tie.
        if is_full(board):
            print("It's a tie!")
            break

        # AI's turn.
        print("AI is making a move...")
        ai_move = get_best_move(board)  # Get the AI's best move.
        board[ai_move] = 'O'  # Place 'O' on the board for the AI's move.
        print_board(board)  # Print the updated board.

        # Check if the AI has won.
        if check_win(board, 'O'):
            print("AI wins!")
            break
        # Check if the board is full, indicating a tie.
        if is_full(board):
            print("It's a tie!")
            break

# Entry point for the program.
if __name__ == "__main__":
    play_game()  # Start the game.
```
